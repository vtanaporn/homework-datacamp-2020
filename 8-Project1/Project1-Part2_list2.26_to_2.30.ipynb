```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import plotly.express as px
import seaborn as sns
%matplotlib inline
```

```python
df = pd.read_csv('AB_NYC_2019.csv')
df
```

```python
df2=df
df2 = df2.dropna()
df2
```

```python
df2['last_review'] = pd.to_datetime(df2['last_review'])
df2
```

### Quiz26 สร้าง Column ใหม่ โดยคำนวณ ณ เวลาปัจจุบันถึงวันที่ Last Review ห่างกันกี่วัน (Hint: ใช้ datetime ปัจจุบัน - datetime last review)

```python
import datetime

today = datetime.datetime.today()
today
```

```python
old_day = df2['last_review'][0]
old_day
```

```python
(today - old_day).days
```

```python
df2['Diff'] = df2['last_review'].apply(lambda past: (today-past).days)
```

```python
df2
```

### Quiz27 สร้าง Pie Chart ของระยะห่างของเวลา ระหว่างปัจจุบันถึง Last review เฉลี่ย ของแต่ละเขต (หน่วยเป็นวัน)

```python
df3=df2.groupby('neighbourhood_group').mean().reset_index()
```

```python
fig = px.pie(df3, values='Diff', names='neighbourhood_group')
fig.show()
```

### Quiz28 สร้าง Bar Plot ของระยะห่างของเวลา ระหว่างปัจจุบันถึง Last review เฉลี่ย ของแต่ละพื้นที่ 10 พื้นที่แรกที่มีระยะเวลามากที่สุด (หน่วยเป็นวัน)

```python
df4 = df2.groupby('neighbourhood').mean().reset_index().sort_values('Diff')[::-1].head(10)
df4
```

```python
fig = plt.figure(figsize=(16,8))
sns.barplot(x='neighbourhood', y='Diff', data=df4)
```

### Quiz29 สร้าง Bar Plot ของค่าเฉลี่ยระยะเวลาห่างระหว่างปัจจุบันถึง Last review เฉลี่ย ของแต่ละพื้นที่ 10 พื้นที่แรกที่มีระยะเวลาน้อยที่สุด

```python
df5 = df2.groupby('neighbourhood').mean().reset_index().sort_values('Diff').head(10)
df5
```

```python
fig = plt.figure(figsize=(16,8))
sns.barplot(x='neighbourhood', y='Diff', data=df5)
```

### Quiz30 สร้าง Scatter Plot หาความสัมพันธ์ระหว่าง ระยะเวลาห่างระหว่างปัจจุบันถึง Last review กับ Minimum Nights (Hint: Correlation)

```python
fig = px.scatter(df2, x='Diff', y='minimum_nights')
fig.show()
```
