### Quiz17 จงเพิ่ม 2 คอลัมน์ชื่อ Year และ Month จาก last review ต่อท้าย โดยใช้ข้อมูลจากคอลัมน์ last_review (Hint: ใช้ lambda)

```python
df2=df
df2['last_review'][0].split('-')[0]
df2 = df2.dropna()
```

```python
df2
```

```python
df2['Year-LastReview'] = df2['last_review'].apply(lambda x: x.split('-')[0])
df2['Month-LastReview'] = df2['last_review'].apply(lambda x: x.split('-')[1])
df2['Day-LastReview'] = df2['last_review'].apply(lambda x: x.split('-')[2])
df2
```
