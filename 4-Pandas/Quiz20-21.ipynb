```python
import numpy as np
import pandas as pd

df = pd.read_csv('train.csv')
df
```

```python
# Check missing values
print("Number of Survived rows = ", df['Survived'].count())
print("Number of Age rows = ", df['Age'].count())
```

```python
# Select only Age and survived column
df2=df[['Age','Survived']]
print('Dataframe select only Age and survived column:')
print(df2)

# Remove rows with null age
df2=df2.dropna(subset=['Age'])
print('After remove Age with null value:')
print(df2)
```

```python
print("Number of passenger group by survival (survived = 1, not survived = 0):")
print(df2['Survived'].value_counts())
print("Minimum age of passenger = ", df2['Age'].min())
print("Maximum age of passenger = ", df2['Age'].max())
```

```python
num_age0to10 = np.sum(df2['Age']<=10)
num_age11to20 = np.sum((df2['Age']>10) & (df2['Age']<=20))
num_age21to30 = np.sum((df2['Age']>20) & (df2['Age']<=30))
num_age31to40 = np.sum((df2['Age']>30) & (df2['Age']<=40))
num_age41to50 = np.sum((df2['Age']>40) & (df2['Age']<=50))
num_age51to60 = np.sum((df2['Age']>50) & (df2['Age']<=60))
num_age61to70 = np.sum((df2['Age']>60) & (df2['Age']<=70))
num_age71to80 = np.sum((df2['Age']>70) & (df2['Age']<=80))
num_agemorethan80 = np.sum(df2['Age']>80)
print("Number of passenger age 0 to 10 = ", num_age0to10)
print("Number of passenger age 11 to 20 = ", num_age11to20)
print("Number of passenger age 21 to 30 = ", num_age21to30)
print("Number of passenger age 31 to 40 = ", num_age31to40)
print("Number of passenger age 41 to 50 = ", num_age41to50)
print("Number of passenger age 51 to 60 = ", num_age51to60)
print("Number of passenger age 61 to 70 = ", num_age61to70)
print("Number of passenger age 71 to 80 = ", num_age71to80)
print("Number of passenger age more than 80 = ", num_agemorethan80)
```

```python
df3=df2[df2['Survived']==1]
print("Data only survive passenger:\n")
print(df3)
```

```python
num_survive_age0to10 = np.sum(df3['Age']<=10)
num_survive_age11to20 = np.sum((df3['Age']>10) & (df3['Age']<=20))
num_survive_age21to30 = np.sum((df3['Age']>20) & (df3['Age']<=30))
num_survive_age31to40 = np.sum((df3['Age']>30) & (df3['Age']<=40))
num_survive_age41to50 = np.sum((df3['Age']>40) & (df3['Age']<=50))
num_survive_age51to60 = np.sum((df3['Age']>50) & (df3['Age']<=60))
num_survive_age61to70 = np.sum((df3['Age']>60) & (df3['Age']<=70))
num_survive_age71to80 = np.sum((df3['Age']>70) & (df3['Age']<=80))
print("Number of survive passenger age 0 to 10 = ", num_survive_age0to10)
print("Number of survive passenger age 11 to 20 = ", num_survive_age11to20)
print("Number of survive passenger age 11 to 20 = ", num_survive_age21to30)
print("Number of survive passenger age 11 to 20 = ", num_survive_age31to40)
print("Number of survive passenger age 11 to 20 = ", num_survive_age41to50)
print("Number of survive passenger age 11 to 20 = ", num_survive_age51to60)
print("Number of survive passenger age 11 to 20 = ", num_survive_age61to70)
print("Number of survive passenger age 11 to 20 = ", num_survive_age71to80)
```

```python
print("Percentage of survival classified by age range:")
print("  0 to 10 = %.3f" %(num_survive_age0to10*100/num_age0to10))
print(" 11 to 20 = %.3f" %(num_survive_age11to20*100/num_age11to20))
print(" 21 to 30 = %.3f" %(num_survive_age21to30*100/num_age21to30))
print(" 31 to 40 = %.3f" %(num_survive_age31to40*100/num_age31to40))
print(" 41 to 50 = %.3f" %(num_survive_age41to50*100/num_age41to50))
print(" 51 to 60 = %.3f" %(num_survive_age51to60*100/num_age51to60))
print(" 61 to 70 = %.3f" %(num_survive_age61to70*100/num_age61to70))
print(" 71 to 80 = %.3f" %(num_survive_age71to80*100/num_age71to80))
```
