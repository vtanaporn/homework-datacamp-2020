### Quiz2 สร้างกราฟฟังก์ชั่น Sin (เหมือนตัวอย่าง) พร้อมใส่ชื่อกราฟ รายละเอียดแต่ละแกน เปลี่ยนสีเส้นเป็นสีชมพู

```python
x = np.arange(0,20)
y = np.sin(x)
plt.plot(x,y, c='#FF33C1')
plt.xlabel('x')
plt.ylabel('y')
plt.title('y = sin(x)')
plt.show()
```

### Quiz3 สร้างกราฟธรรมดาเป็นฟังก์ชั่น X = Y^2 พร้อมใส่ชื่อกราฟ รายละเอียดแต่ละแกน เปลี่ยนสีเส้นเป็นสีเขียวอ่อนโดยใช้ค่ารหัสสีฐานสิบหก

```python
y = np.arange(0,20, 2)
x = y**2
print("x: ", x)
print("y: ", y)
plt.plot(x,y, c='#B8FF33')
plt.xlabel('x')
plt.ylabel('y')
plt.title('x = y^2')
plt.show()
```

### Quiz4 สร้างกราฟฟังก์ชั่น Cos (เหมือนตัวอย่าง) พร้อมใส่ชื่อกราฟ รายละเอียดแต่ละแกน เปลี่ยนสีเส้นเป็นสีแดง ใส่ Line Style เป็นเส้น - -

### Quiz5 สร้างกราฟธรรมดาเป็นฟังก์ชั่น Y = X^2 และ Y = X^3 พล็อตลงในรูปเดียวกัน พร้อมใส่ชื่อกราฟ รายละเอียดแต่ละแกน ใช้สีแดงและน้ำเงินตามลำดับ โดยเส้นแรกใช้วงกลมเป็น marker และเส้นสองใช้สามเหลี่ยมเป็น marker

```python
x = np.arange(0,20)
y1 = x**2
y2 = x**3
plt.plot(x, y1, c='r', marker='o')
plt.plot(x, y2, c='b', marker='^')
plt.xlabel('x')
plt.ylabel('y')
plt.title('x=y^2 vs. x=y^3')
plt.show()
```

### Quiz6 สร้างกราฟธรรมดาเป็นฟังก์ชั่น Y = X^2 และ X = Y^2 พล็อตลงในรูปเดียวกัน พร้อมใส่ชื่อกราฟ รายละเอียดแต่ละแกน ใช้สีแดงและน้ำเงินตามลำดับ โดยเส้นแรกใช้วงกลมเป็น marker และเส้นสองใช้สามเหลี่ยมเป็น marker และใส่ legend และเพิ่มเส้นตารางข้างหลัง

```python
x = np.arange(0,20)
y1 = x**2
y2 = np.sqrt(x)
plt.plot(x, y1, label='y=x^2', c='r', marker='o')
plt.plot(x, y2, label='x=y^2', c='b', marker='^')
plt.xlabel('x')
plt.ylabel('y')
plt.title('x=y^2 VS. y=x^2')
plt.grid()
plt.legend(loc=0)
plt.show()
```

### Quiz7 สร้างกราฟธรรมดาเป็นฟังก์ชั่น X = Y^2 + 4*Y พร้อมใส่ชื่อกราฟ รายละเอียดแต่ละแกน เปลี่ยนสีเส้นเป็นสีเหลืองอ่อนโดยใช้ค่ารหัสสีฐานสิบหก จำกัดแกน X และ Y ให้เหมาะสม

```python
y = np.arange(0, 20)
x = y**2 + 4*y
print("x: ", x)
print("y: ", y)
plt.plot(x, y, linewidth=3, c='#F5E94E')
plt.xlabel('x')
plt.ylabel('y')
plt.title('x = y^2 + 4x')
plt.xlim([0,100])
plt.ylim([0,8])
plt.show()
```

### Quiz9 ให้หาอัตราส่วนตั๋วโดยสารชั้น 1 2 3 และพล็อตเป็น Pie chart

```python
df['Pclass'].value_counts()
```

```python
num_ticket_class1 = np.sum(df['Pclass']==1)
num_ticket_class2 = np.sum(df['Pclass']==2)
num_ticket_class3 = np.sum(df['Pclass']==3)

num_passenger = df['Pclass'].count()

ratio_ticket_class1 = num_ticket_class1*100/num_passenger
ratio_ticket_class2 = num_ticket_class2*100/num_passenger
ratio_ticket_class3 = num_ticket_class3*100/num_passenger

print("Number of ticket class 1 = ", num_ticket_class1)
print("Number of ticket class 2 = ", num_ticket_class2)
print("Number of ticket class 3 = ", num_ticket_class3)
print("\nNumber of all passenger = ", num_passenger)

print("\nRatio of ticket class 1 = ", ratio_ticket_class1)
print("Ratio of ticket class 2 = ", ratio_ticket_class2)
print("Ratio of ticket class 3 = ", ratio_ticket_class3)
```

```python
plt.figure(figsize=[5,5])
labels = ['Class 1', 'Class 2', 'Class 3']
plt.pie([ratio_ticket_class1, ratio_ticket_class2, ratio_ticket_class3], labels=labels)
plt.title('Ratio of ticket class')
plt.show()
```

### Quiz10 ให้หาอัตราส่วนของจุดหมายปลายทาง (Embark) และพล็อต Pie chat พร้อมทั้งให้ Queenstown เป็น explode

```python
df['Embarked'].value_counts()
```

```python
num_embark_s = np.sum(df['Embarked']=='S')
num_embark_c = np.sum(df['Embarked']=='C')
num_embark_q = np.sum(df['Embarked']=='Q')

total_embark = df['Embarked'].count()

ratio_embark_s = num_embark_s*100/total_embark
ratio_embark_c = num_embark_c*100/total_embark
ratio_embark_q = num_embark_q*100/total_embark

print("Number of embark at Southamton = ", num_embark_s)
print("Number of embark at Cherbourg = ", num_embark_c)
print("Number of embark at Queenstown = ", num_embark_q)
print(total_embark)
print("Ratio of embark at Southamton = ", ratio_embark_s)
print("Ratio of embark at Cherbourg = ", ratio_embark_c)
print("Ratio of embark at Queenstown = ", ratio_embark_q)
```

```python
plt.figure(figsize=[5,5])
explode = [0,0,0.1]
labels = ['Southamton', 'Cherbourg', 'Queenstown']
plt.pie([ratio_embark_s, ratio_embark_c, ratio_embark_q], labels=labels, explode=explode)
plt.title('Ratio of embark')
plt.show()
```
