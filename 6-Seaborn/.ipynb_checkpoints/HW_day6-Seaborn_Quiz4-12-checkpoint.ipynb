```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

## ให้ import ไฟล์ titanic ไฟล์เดิมจากบทเรียน Pandas เข้าสู่ Jupyter Notebook สามารถใช้ Pandas มาร่วมทำด้วย

```python
df = pd.read_csv('train.csv')
df
```

### Quiz 4 สร้าง count plot นับข้อมูล Embarked โดยสร้างกราฟแบบแนวตั้งและแนวนอน

```python
# Vertical
sns.countplot(x='Embarked', data=df)
```

```python
# Horizontal
sns.countplot(y='Embarked', data=df)
```

### Quiz 5 สร้าง count plot นับข้อมูล Pclass โดยสร้างกราฟแบบแนวตั้งและแนวนอน

```python
# Vertical
sns.countplot(x='Pclass', data=df)
```

```python
# Horizontal
sns.countplot(y='Pclass', data=df)
```

### Quiz 6 สร้าง box plot ให้แกน X เป็นเพศ และแกน Y เป็นราคา

```python
fig = plt.figure(figsize=(5,10))
sns.boxplot(x='Sex', y='Fare', data=df)
```

### Quiz 7 สร้าง box plot ให้แกน X เป็นPclass และแกน Y เป็นราคา

```python
fig = plt.figure(figsize=(5,20))
sns.boxplot(x='Pclass', y='Fare', data=df)
```

### Quiz 8 สร้าง barplot โดยให้ แกน X เป็นจุดหมายปลายทางและแกน Y เป็นราคา

```python
sns.barplot(x='Embarked', y='Fare', data=df)
```

### Quiz 9 สร้าง barplot โดยให้ แกน X เป็นPclass และแกน Y เป็นอายุ

```python
sns.barplot(x='Pclass', y='Age', data=df)
```

### Quiz 10 สร้าง Stripplot โดยให้แกน X เป็น Pclass และ แกน Y เป็นราคา

```python
sns.stripplot(x='Pclass', y='Fare', data=df)
```

### Quiz 11 สร้าง Stripplot โดยให้ แกน X เป็น Survived แกน Y เป็นอายุ และ hue เป็นเพศ

```python
fig = plt.figure(figsize=(5,10))
sns.stripplot(x='Pclass', y='Fare', data=df, hue="Sex")
```

### Quiz 12 สร้าง Swarmplot โดยให้ แกน X เป็นเพศและแกน Y เป็นอายุ

```python
sns.swarmplot(x='Sex', y='Age', data=df)
```
