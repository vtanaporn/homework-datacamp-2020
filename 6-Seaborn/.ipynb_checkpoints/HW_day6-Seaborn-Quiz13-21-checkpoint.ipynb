```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

#### ให้ import ไฟล์ titanic ไฟล์เดิมจากบทเรียน Pandas เข้าสู่ Jupyter Notebook สามารถใช้ Pandas มาร่วมทำด้วย

```python
df = pd.read_csv('train.csv')
df
```

### Quiz 13 สร้าง HeatMap โดย crosstab กำหนดให้ index เป็น Pclass และ column เป็นเพศ

```python
df_crosstab_ClassSex = pd.crosstab(df['Pclass'], df['Sex'])
df_crosstab_ClassSex
```

```python
sns.heatmap(df_crosstab_ClassSex)
```

### Quiz 14 สร้าง HeatMap โดย pivot table กำหนดให้ index เป็น Pclass และ column เป็นเพศ และ values เป็น ราคา

```python
df_pivotTbl_FareOnClassAndSex = df.pivot_table(index='Pclass', columns='Sex', values='Fare')
df_pivotTbl_FareOnClassAndSex
```

```python
sns.heatmap(df_pivotTbl_FareOnClassAndSex, annot=True, cmap='coolwarm')
```

### Quiz15 สร้าง HeatMap โดย pivot table กำหนดให้ index เป็น Pclass และ column เป็นเพศ และ values เป็น ราคา และกำหนดเส้นสีขาวขนาด 2 เป็นเส้นแบ่ง

```python
sns.heatmap(df_pivotTbl_FareOnClassAndSex, annot=True, cmap='coolwarm', linecolor='white', linewidth=2)
```

### Quiz 16 สร้าง Heatmap โดย Correlation ของทั้ง DataFrame

```python
df_corr = df.corr()
df_corr
```

```python
sns.heatmap(df_corr, annot=True, cmap='coolwarm', linecolor='white', linewidth=2)
```

### Quiz17 สร้าง Clustermap ของทั้ง DataFrame

```python
sns.clustermap(df_corr, annot=True, cmap='coolwarm', linecolor='white', linewidth=2)
```
